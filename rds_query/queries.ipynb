```python
from SPARQLWrapper import SPARQLWrapper, JSON, CSV
import io
import pandas as pd
sparql = SPARQLWrapper("http://localhost:3030/graph")

package = input('Package you want to find')
version = input('version')
if version == '':
  sparql.setQuery("""
  PREFIX sd: <https://w3id.org/okn/o/sd#>
  prefix c2t: <https://w3id.org/c2t#> 
  prefix c2ti: <https://w3id.org/c2t/i/> 
  prefix dc: <http://purl.org/dc/elements/1.1/> 
  prefix dpv: <http://dockerpedia.inf.utfsm.cl/vocab#> 
  prefix owl: <http://www.w3.org/2002/07/owl#> 
  prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#> 
  prefix rr: <http://www.w3.org/ns/r2rml#> 
  prefix rml: <http://semweb.mmlab.be/ns/rml#> 
  prefix ql: <http://semweb.mmlab.be/ns/ql#> 
  prefix : <http://example.org/> 


  SELECT ?tag
  WHERE {
    ?o a dpv:PackageVersion.
    ?o sd:name '"""+package+"""'.
    ?o dpv:isInstalledOn ?image.
    ?image dpv:tag ?tag
  } 
  """)
else:
  sparql.setQuery("""
  PREFIX sd: <https://w3id.org/okn/o/sd#>
  prefix c2t: <https://w3id.org/c2t#> 
  prefix c2ti: <https://w3id.org/c2t/i/> 
  prefix dc: <http://purl.org/dc/elements/1.1/> 
  prefix dpv: <http://dockerpedia.inf.utfsm.cl/vocab#> 
  prefix owl: <http://www.w3.org/2002/07/owl#> 
  prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#> 
  prefix rr: <http://www.w3.org/ns/r2rml#> 
  prefix rml: <http://semweb.mmlab.be/ns/rml#> 
  prefix ql: <http://semweb.mmlab.be/ns/ql#> 
  prefix : <http://example.org/> 


  SELECT ?tag
  WHERE {
    ?o a dpv:PackageVersion.
    ?o sd:name '"""+package+"""'.
    ?o sd:hasVersionId '"""+version+"""'.
    ?o dpv:isInstalledOn ?image.
    ?image dpv:tag ?tag
  } 
  """)
sparql.setReturnFormat(CSV)
results_lib = sparql.query().convert()
results_2lib = io.StringIO(results_lib.decode('utf-8'))
lib_results = pd.read_csv(results_2lib)
print(lib_results)
```

```python
image = input('Image')
sparql.setQuery("""
PREFIX sd: <https://w3id.org/okn/o/sd#>
prefix c2t: <https://w3id.org/c2t#> 
prefix c2ti: <https://w3id.org/c2t/i/> 
prefix dc: <http://purl.org/dc/elements/1.1/> 
prefix dpv: <http://dockerpedia.inf.utfsm.cl/vocab#> 
prefix owl: <http://www.w3.org/2002/07/owl#> 
prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#> 
prefix rr: <http://www.w3.org/ns/r2rml#> 
prefix rml: <http://semweb.mmlab.be/ns/rml#> 
prefix ql: <http://semweb.mmlab.be/ns/ql#> 
prefix dct: <https://purl.org/dc/terms/> 
prefix : <http://example.org/> 

SELECT ?image  ?tag ?architecture ?size ?osDescription 
WHERE { ?image a dpv:Image.
  ?image dpv:tag '"""+image+"""'.
  ?image dpv:tag ?tag.
  ?os c2t:isOperativeSystemOf ?image .
  ?image c2t:architecture ?architecture .
  ?image dpv:size ?size .
  ?os sd:description ?osDescription .
  
} 
""")
results_im = sparql.query().convert()
results_2im = io.StringIO(results_im.decode('utf-8'))
im_results = pd.read_csv(results_2im)
print(im_results)


sparql.setQuery("""
PREFIX sd: <https://w3id.org/okn/o/sd#>
prefix c2t: <https://w3id.org/c2t#> 
prefix c2ti: <https://w3id.org/c2t/i/> 
prefix dc: <http://purl.org/dc/elements/1.1/> 
prefix dpv: <http://dockerpedia.inf.utfsm.cl/vocab#> 
prefix owl: <http://www.w3.org/2002/07/owl#> 
prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#>
prefix rr: <http://www.w3.org/ns/r2rml#> 
prefix rml: <http://semweb.mmlab.be/ns/rml#> 
prefix ql: <http://semweb.mmlab.be/ns/ql#> 
prefix dct: <https://purl.org/dc/terms/> 
prefix : <http://example.org/> 

SELECT ?packageT (COUNT(?package) as ?total)
WHERE {
    ?image a dpv:Image.
    ?image dpv:tag '"""+image+"""'.
	?package dpv:isInstalledOn ?image;
			c2t:operativeSystemPackageVersion ?packageT .
}GROUP BY (?packageT)
""")
results_pa = sparql.query().convert()
results_2pa = io.StringIO(results_pa.decode('utf-8'))
pa_results = pd.read_csv(results_2pa)
print(pa_results)
```
